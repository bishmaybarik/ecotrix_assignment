```python
import geopandas as gpd

# Path to the SHP file
shp_file = '/Users/bishmaybarik/Library/CloudStorage/OneDrive-ShivNadarInstitutionofEminence/Discrimination_Assignment/02. Output/Shapefile/TN_Shapefile/tamil_nadu.shp'

# Load the Shapefile into a GeoDataFrame (GDF)
gdf = gpd.read_file(shp_file)

# Extract full coordinates of each geometry (Polygon or LineString)
gdf['full_coordinates'] = gdf['geometry'].apply(
    lambda geom: list(geom.exterior.coords) if geom.geom_type == 'Polygon' 
                 else list(geom.coords) if geom.geom_type in ['LineString', 'Point'] 
                 else None
)

# View all the coordinates
print(gdf[['geometry', 'full_coordinates']])  # Show the geometries and their full coordinates
```
