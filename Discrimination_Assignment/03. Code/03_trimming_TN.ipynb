```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Path to the shapefile
shapefile_path = '/Users/bishmaybarik/Library/CloudStorage/OneDrive-ShivNadarInstitutionofEminence/Discrimination_Assignment/02. Output/Shapefile/DISTRICT_BOUNDARY/DISTRICT_BOUNDARY.shp'

# Load the shapefile
districts = gpd.read_file(shapefile_path)

# Filter for only 'Tamil Nadu' (with encoding issue) in the 'STATE' column
tamil_nadu_districts = districts[districts['STATE'] == 'TAMIL N>DU']

# Plot the filtered shapefile
fig, ax = plt.subplots(figsize=(10, 10))
tamil_nadu_districts.plot(ax=ax, color='lightgreen', edgecolor='black')
ax.set_title('District Boundaries in Tamil Nadu')
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Path to the original shapefile
shapefile_path = '/Users/bishmaybarik/Library/CloudStorage/OneDrive-ShivNadarInstitutionofEminence/Discrimination_Assignment/02. Output/Shapefile/DISTRICT_BOUNDARY/DISTRICT_BOUNDARY.shp'

# Load the shapefile
districts = gpd.read_file(shapefile_path)

# Filter for only 'Tamil Nadu' in the 'STATE' column
tamil_nadu_districts = districts[districts['STATE'] == 'TAMIL N>DU']

# Save the filtered shapefile
output_path = '/Users/bishmaybarik/Library/CloudStorage/OneDrive-ShivNadarInstitutionofEminence/Discrimination_Assignment/02. Output/Shapefile/DISTRICT_BOUNDARY/tamilnadu_final.shp'
tamil_nadu_districts.to_file(output_path)

# Plot to confirm
fig, ax = plt.subplots(figsize=(10, 10))
tamil_nadu_districts.plot(ax=ax, color='lightgreen', edgecolor='black')
ax.set_title('District Boundaries in Tamil Nadu')
plt.show()

print(f"Filtered shapefile saved to: {output_path}")
```
